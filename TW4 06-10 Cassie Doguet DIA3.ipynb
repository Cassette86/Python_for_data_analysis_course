## Exercise 2 [Matplotlib's Object Oriented API]
### Matplotlib have an object-oriented API used to create more complex figures. Code is a little more complicated, but the advantage is that we now have full control of where the plot axes.  
### Create the same chart as 1.1 using a new method: instantiate figure objects and then call methods or attributes from that object.
#### 1. Create the figure below. (The blue curve represents y = x2, while the red curve represents x = y2)

```python
x=np.linspace(0,5,11)
y = x**2
plt.rcParams["figure.figsize"] = [10, 7]
plt.rcParams["figure.autolayout"] = True
fig, ax1 = plt.subplots()

ax1.plot(y,x,color='pink',marker='o')
ax1.set_title("Global title")
ax1.set_xlabel("X label original")
ax1.set_ylabel("Y label original")

ax2 = fig.add_axes([0.45, 0.20, 0.5, 0.25])
ax2.plot(x,y,'c')
ax2.set_xlabel("X label nested")
ax2.set_ylabel("Y label nested")
ax2.set_title("Nested title")

plt.show()
```

#### 2. Explore colors and markers style. Create a new chart, such as x goes from 0 to 10, and  y = x + index. Index is the number of curves you are going to draw. For example, the  first curve will match y = x +0, the second y = x + 1, etc. Create multiple combinations of different colors, markers & marker size in this  graphic. Here’s an example using three curves:

```python
x = np.linspace(0,5,11)

colors = ['b', 'r', 'k']
markers = ['', 'o', '+']

plt.figure(figsize=(8, 6))

for i in range(3):
    y = x + i
    plt.plot(x, y, label=f'Curve {i}', color=colors[i], marker=markers[i])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Curves with Different Styles')
plt.legend()

plt.show()
```

#### 1. Use the dataframe you created on 2.25, representing the relationship between ‘moving obstacles’ and ‘transport vehicles’. Create a dataframe using this data, marking the correlation between each variable in the heatmap. Add a title and change the style of this plot. Here’s an example of the expected output:

```python
import seaborn as sns

df = pd.read_csv('vehicules_2016.csv', sep=',')
df.columns = ["Num_Acc", "sens_de_Circulation", "catV","nb_occupants", "obstacle_fixe", "obstacle_mobile","choc", "manoeuvre" , "num_vehicule"]
array_obstacle_fixe=np.array(df['obstacle_fixe'])
array_obstacle_fixe = np.nan_to_num(array_obstacle_fixe)
from mapping import *
df['catV'] = df['catV'].map(info_vehicules)
df['obstacle_fixe'] = df['obstacle_fixe'].map(obstacle_fixe_labels)
df['obstacle_mobile'] = df['obstacle_mobile'].map(obstacle_labels)
```

```python
df2=df.groupby(['catV','obstacle_mobile'])['nb_occupants'].count().reset_index()

df3=df2.pivot(index='catV',columns='obstacle_mobile',values='nb_occupants')

public_transport_list=['Autobus','Autocar','Train','Tramway']
df3=df3.loc[public_transport_list]

df3=df3.dropna(axis=1,how='all')
sns.heatmap(df3, annot=True)
```

```python
df3['Total'] = df3.sum(axis=1)
df3.head()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df3['Total'], labels=df3.index, autopct='%1.1f%%', startangle=140)
plt.title('Répartition des valeurs par véhicule')
plt.axis('equal')
plt.show()
```
