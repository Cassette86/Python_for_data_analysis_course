- f(x) = sinx * cosx
- f(x) = cosx**2 - sinx**2

#### Pixelize Mona Lisa's image on 3 different levels, and explain the result

```python
num_levels = 3

height, width = 1200,800
region_height = height // num_levels
region_width = width // num_levels

pixelized_image = np.zeros_like(array_npy)

for i in range(num_levels):
    for j in range(num_levels):
        y_start = i * region_height
        y_end = (i + 1) * region_height
        x_start = j * region_width
        x_end = (j + 1) * region_width
        region = array_npy[y_start:y_end, x_start:x_end]
        average_color = np.mean(region)
        pixelized_image[y_start:y_end, x_start:x_end] = average_color

plt.imshow(pixelized_image) 
plt.axis('off')
plt.show()
```

### Transformation - Use the monlisa.txt file
#### Create a filter on the image to have a darker Mona Lisa

```python
with open('Datasets/monalisa.txt', 'r') as file:
    lines = file.readlines()

array = np.array([list(map(int, line.strip().split())) for line in lines])
darkened_array = array * 0.5
darkened_array = np.clip(darkened_array, 0, 255).astype(np.uint8)

plt.imshow(darkened_array)
```

#### Create a vertical gradient from right to left allowing the image to be darker on the left

```python
image_height = 200
image_width = 134

image_height, image_width = array.shape

for x in range(image_width):
    gradient_value = int(255 * (x / image_width))
    array[:, x] = gradient_value
    
plt.imshow(array)
```

#### Create a horizontal gradient that will make the Mona Lisa darker on top

```python
with open('Datasets/monalisa.txt', 'r') as file:
    lines = file.readlines()

array = np.array([list(map(int, line.strip().split())) for line in lines])

image_height, image_width = array.shape

for y in range(image_height):
    darkening_factor = y / image_height
    array[y, :] = (array[y, :] * (1 - darkening_factor)).astype(int)
    
plt.imshow(array)
```
