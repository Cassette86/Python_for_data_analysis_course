#### Map the vehicle category, obstacles (fixed and mobile) using the dictionary defined in mapping.py (you can use the mapping.py file, but you can also create your own mappings).


```python
import sys
sys.path.append('Datasets/')
from mapping import info_vehicules, obstacle_labels, obstacle_fixe_labels

df['catV'] = df['catV'].map(info_vehicules)
df['obstacle_fixe'] = df['obstacle_fixe'].map(obstacle_fixe_labels)
df['obstacle_mobile'] = df['obstacle_mobile'].map(obstacle_labels)
```

```python
df.head()
```

#### Create a new database, projecting the category of the vehicle in relation to the moving obstacle. The values of the data frame will be the sum of the respective occupant numbers. This database can be referred as a heatmap of the number of occupants by type of transport and obstacle.

```python
df_bis=df.groupby(['catV','obstacle_mobile'])['nb_occupants'].sum()
df_bis.head()
```

#### Using the previous results, obtain the most dangerous mode of transport for the public

```python
df_bis=df_bis.reset_index()
df_bis=df_bis.set_index(['catV','obstacle_mobile'])

public_transport_list=['Autobus','Autocar','Train','Tramway']
df_bis.loc[public_transport_list].groupby(['catV'])['nb_occupants'].sum().sort_values(ascending=False)
```

#### Sort the heatmap from the most dangerous to the least dangerous means of transport

```python
import seaborn as sns

heatmap_data=df_bis.pivot_table(index=['catV'], values=['nb_occupants'], aggfunc=np.sum)
heatmap_data=heatmap_data.sort_values(by=['nb_occupants'], ascending=False)
sns.heatmap(heatmap_data, annot=True)
```

#### Get the most dangerous moving obstacle when using public transportation

```python
df_bis.loc[public_transport_list].groupby(['obstacle_mobile'])['nb_occupants'].sum().sort_values(ascending=False).head(1)
```

#### Sort the heatmap from the most dangerous to the least dangerous entity (means of transport, and moving obstacle)

```python
df_bis=df_bis.reset_index()
df_bis=df_bis.set_index(['catV','obstacle_mobile'])
df_bis=df_bis.loc[public_transport_list].sort_values(by=['nb_occupants'], ascending=False)
sns.heatmap(df_bis, annot=True)
```
